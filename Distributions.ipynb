# Distriutions of the features

In this section the the distributions and boxplot for all the features were build (first for numerical ones, then for categorical). For every feature there is original histogram and boxplot and then the plot without sum outliers when needed.

```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set()
sns.set_context('paper')

FILE_PATH = 'C:/Projects/BatteryLevelAnalysis/data/battery_analysis.csv'
df = pd.read_csv(FILE_PATH)
```
